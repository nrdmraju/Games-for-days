```python
# Read video game sales data scrapped from https://www.vgchartz.com/gamedb/
file = '../raw_data/vgsales-12-4-2019.csv'
df = pd.read_csv(file)
df
```

```python
#  view of the range of the video game critic scores
# df['Vgchartzscore'].value_counts()
```

```python
# fill NaN values in the Total_Shipped and Global Sales columns to 0 - so we can add these values in next step
df['Total_Shipped'].fillna(0, inplace=True)
df['Global_Sales'].fillna(0, inplace=True)
```

```python
# total shipped and total sales are measured in the same units (per million units sold. it is not a currency value)
# create new total column so the video game sales data(to have less 0 values in each row)
# this was done so there is more sales data we can anaylyze 

df['Global_Total'] = df['Total_Shipped']+df['Global_Sales']
```

```python
# create view of the non null values in the new Global_Total column to make sure previous worked

#df[df['Total_Shipped'].notnull()]
#df[df['Global_Sales'].notnull()]
df[df['Global_Total'].notnull()]
```

```python
# identify rows in the dataframe where Global_Total is equal zero, rows we will need to drop
series = df[df['Global_Total']==0].index
series
```

```python
# df = df.dropna(axis=0, subset=['Year','NA_Sales','PAL_Sales','JP_Sales'])
# df = df.dropna(axis=0, subset=['Total_Shipped'])
```

```python
# drop the rows where Global Tatal is equal to zero and create new dataframe
clean_df = df.drop(df.index[series])
clean_df
```
